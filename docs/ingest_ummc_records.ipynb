````python
from json import dumps, loads
from getpass import getpass
from smtplib import SMTP_SSL
from ssl import create_default_context
from datetime import datetime, timedelta

from requests import post as POST, get as GET
from osgeo import ogr
import numpy as np

def get_echo_token(client: str="CMRCollectionsWorkflow-jmcnelis"):
    """Returns a temporary Echo-Token following URS authentication.

    Parameters
    ----------
    url (str, optional): the target CMR endpoint to POST the XML data
    headers: 
    client (str, optional): the client/application name
    
    Returns
    -------
    str: a temporary Echo-Token for interacting with CMR API
    
    """
    
    # Make dict of user data for token API. Info will be purged on function close.
    data = {

        # Prompt the user for their URS/Earthdata username:
        'username': input("Username: "),

        # Securely prompt the user for their password:
        'password': getpass("Password: "),

        # Provide a string identifier for the host client/application:
        'client_id': client,

        # Get the host's external IP address:
        'user_ip_address': "128.149.85.238",

    }

    # Format the XML data to the string required to get the user's XML token.
    xml = f'<token>{"".join([f"<{t}>{v}</{t}>" for t,v in data.items()])}</token>'

    # Request the token and get the response text.
    response = POST(
        url=f"https://{cmr_domain}/legacy-services/rest/tokens", 
        headers={'Content-Type': "application/xml"}, 
        data=xml,
    ).text

    # Split the response text. 'IndexError' results from a bad request.
    try:
        token = response.split("<id>")[1].split("</id")[0]
    except IndexError as e:
        raise e
    else:
        return token
    
token = get_echo_token()
token
````

````python
def get_nid(x):
    x = x['EntryTitle']
    return quote_plus(x)

def get_rec(x):
    with open(x, "r") as f:
        return loads("".join([l.strip() for l in f.readlines()]))

test0 = get_rec(ummc_records[0])
print(get_nid(test0))
````

````python
def validate(record: dict):
    nid = get_nid(record)
    try:
        r = put(
            f"https://{cmr_domain}/ingest/providers/POCLOUD/validate/collection/{nid}", 
            headers={'Content-Type': "application/vnd.nasa.cmr.umm+json", 'Echo-Token': token},
            data=dumps(record)
        )
    except Exception as e:
        raise e
    else:
        return r.status_code
````

````python
def get_ingest_cmd(record: str):
    data = dumps(record).replace("'", "\\'")
    return (
        "curl -s -i -XPUT "
        "-H 'Content-Type: application/vnd.nasa.cmr.umm+json' "
        f"-H 'Echo-Token: {token}' "
        f"--data-binary '{data}' "
        f"https://{cmr_domain}/ingest/providers/{cmr_provider}/collections/{get_nid(record)} "
    )

get_ingest_cmd(test0)
````

````python
rtime = datetime.utcnow().strftime('%Y%m%dT%H%M%S')
rresp = abspath(f"../scripts")

with open(f"{rresp}/{rtime}-ummc-ingest.sh", "a") as f:
    f.write("#!/bin/bash \n")

for ix, row in dstab.iterrows():
    rpath = f"../metadata/ummc/{row['ORIGINAL.SHORT_NAME']}.json"
    with open(rpath, "r") as f:    
        rdata = load(f)
    cmd = get_ingest_cmd(rdata)
    with open(f"{rresp}/{rtime}-ummc-ingest.sh", "a") as f:
        f.write(f"# {rdata['ShortName']} \n")
        f.write(cmd + f" >> {rresp}/{rtime}-ummc-ingest.log \n")
````
