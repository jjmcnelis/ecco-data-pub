````python
def _CollectionCitations(x):
    title = x['DATASET.LONG_NAME']
    doi = x['DATASET.PERSISTENT_ID'].replace("PODAAC-","10.5067/")
    url = "https://podaac.jpl.nasa.gov/ECCO"
    return [{'OtherCitationDetails': f"ECCO, Ian Fenty, 2020, {title}, {doi}, {url}"}]

metamap['CollectionCitations'] = dstab.apply(_CollectionCitations, axis=1)
metamap['CollectionCitations']
````

````python
def _AdditionalAttributes(x):
    return [ {
        "Value": x['DATASET.LONG_NAME'],
        "Name" : "Series Name",
        "Description" : "Dataset citation series name",
        "DataType" : "STRING"
      } ]

metamap['AdditionalAttributes'] = dstab.apply(_AdditionalAttributes, axis=1)
metamap['AdditionalAttributes']
````

````python
def _TemporalExtents(x):
    return [
        {
          "EndsAtPresentFlag": False,
          "RangeDateTimes": [
            {
              "BeginningDateTime": "1992-01-01T00:00:00.000Z",
              "EndingDateTime": "2017-12-31T00:00:00.000Z"
            }
          ]
        }
    ]

metamap['TemporalExtents'] = dstab.apply(lambda x: _TemporalExtents(x), axis=1)
metamap['TemporalExtents']
````

````python
def _TemporalKeywords(x):
    if "Daily" in x['DATASET.LONG_NAME']:
        return ["Daily - < Weekly"]
    elif "Monthly" in x['DATASET.LONG_NAME']:
        return ["Monthly - < Annual"]
    else:
        return ["< 1 second"]
    
metamap['TemporalKeywords'] = dstab.apply(_TemporalKeywords, axis=1)
metamap['TemporalKeywords']
````

````python
from jsonschema import RefResolver, Draft7Validator, validate as jsonschema_validate
from jsonschema.exceptions import ValidationError as jsonschema_ValidationError

umm_repo = "https://git.earthdata.nasa.gov/projects/EMFD/repos/unified-metadata-model"

ummc_schema = get(f"{umm_repo}/raw/collection/v1.9/umm-c-json-schema.json").json()
ummcmn_schema = get(f"{umm_repo}/raw/collection/v1.9/umm-cmn-json-schema.json").json()

ummc_schema['$id'] = "umm-c-json-schema.json"
ummcmn_schema['$id'] = "umm-cmn-json-schema.json"

schema_store = {
    ummcmn_schema['$id'] : ummcmn_schema,
    ummc_schema['$id'] : ummc_schema,
}

resolver = RefResolver.from_schema(ummcmn_schema, store=schema_store)
validator = Draft7Validator(ummc_schema, resolver=resolver)
````

````python
def __validate__(r):
    try:
        validator.validate(r)
    except Exception as e:
        print(f"INVALID:\n {e.schema_path}\n {e.message}")
        raise e
    else:
        print("VALID")


__validate__(ecco_ummc_records[0])
````
